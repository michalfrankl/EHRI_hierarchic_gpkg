```python
import geopandas as gpd
import os

input_file = 'test-data/HGC-borders-19380430-v1.gpkg'
core_file_name = os.path.splitext(os.path.basename(input_file))[0]
print(core_file_name)
```

```python
dataset = gpd.read_file(input_file)
dataset.head()
```

```python
for i in range(len(dataset)):
    print(i)
    new_layer_name = core_file_name + "-" + str(i)
    newgdf = dataset.filter(items=[i], axis=0)
    newgdf.info()
    newgdf.to_file(input_file, layer=new_layer_name, driver="GPKG")
```

```python
import fiona
```

```python
for layername in fiona.listlayers(input_file):
    with fiona.open(input_file, layer=layername) as src:
        print(layername, len(src))
```
